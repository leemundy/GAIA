```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import pylab
```

```python
from astropy.table import Table
t = Table.read('TGAS_d125_Wright03_2Mass', format='votable')
print(t.colnames)
import copy
import numpy.ma as ma
Jmag = t['Jmag']
J_mag = (copy.copy(Jmag))
Gmag = t['__Gmag_']
G_mag = np.array(copy.copy(Gmag))
Hmag = t['Hmag']
H_mag = np.array(copy.copy(Hmag))
Kmag = t['Kmag']
K_mag = np.array(copy.copy(Kmag))
Qfl=t['Qfl']
Flag = (copy.copy(t['Qfl']))
eJmag = t['e_Jmag']
eKmag = t['e_Kmag']
eHmag = t['e_Hmag']

print(Jmag.mask[0:6])

for i in range(0,len(Qfl)):
    if(Qfl[i][0] == ord("A")):
        Jmag.mask[i] = False
    else:
        Jmag.mask[i] = True
    if(Qfl[i][1] == ord("A")):
        Hmag.mask[i] = False
    else:
        Hmag.mask[i] = True
    if(Qfl[i][2] == ord("A")):
        Kmag.mask[i] = False
    else:
        Kmag.mask[i] = True
print(Qfl[0:6])
print(Jmag.mask[0:6])

GJ = Gmag - Jmag
GK = Gmag - Kmag
GH = Gmag - Hmag
JK = Jmag - Kmag
HK = Hmag - Kmag
JG = Jmag - Gmag
plt.title("G vs G-J")
plt.xlabel("G-J Magnitude")
plt.ylabel("G Magnitude")
plt.scatter(GJ,Gmag,s=1.0,c='r')
plt.show()

```

```python


plt.title("Error in JMag")
plt.xlabel("J Magnitude")
plt.ylabel("error in J Magnitude")
plt.scatter(Jmag,eJmag,s=1.0,c='r')
plt.show()
```

```python
plt.title("Error in KMag")
plt.xlabel("K Magnitude")
plt.ylabel("error in K Magnitude")
plt.scatter(Kmag,eKmag,s=1.0,c='r')
plt.show()
```

```python
plt.title("Error in HMag")
plt.xlabel("H Magnitude")
plt.ylabel("error in H Magnitude")
plt.scatter(Hmag,eHmag,s=1.0,c='r')
plt.show()
```

```python
plt.title("J-G vs H-K")
plt.xlabel("J-G Magnitude")
plt.ylabel("H-K Magnitude")
plt.scatter(JG,HK,s=1.0,c='r')
plt.show()
```
