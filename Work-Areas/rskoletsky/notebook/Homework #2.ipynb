```python
from astropy.table import Table
import matplotlib.pyplot as plt
import math
import numpy
import time
import sklearn
import scipy
from numpy import loadtxt
```

```python
Lambda=4.455e15
G=6.674e-11
M=1.988e30
e=numpy.linspace(0,10,1000)
mu=G*M
c=(Lambda*Lambda)/mu
one=numpy.ones([1,1000])
w=one-numpy.square(e)
x=numpy.power(w,-1)
A=c*x
```

```python
A=numpy.reshape(A,[1,1000])
e=numpy.reshape(e,[1,1000])
plt.rcParams['figure.figsize']=10,10
plt.plot(e,A,marker='.',color='xkcd:blue',markersize=2)
plt.plot(0.0167086,1.496e11,marker='*',color='xkcd:green',markersize=15)
plt.xlabel('Eccentricity e')
plt.ylabel('Semimajor Axis A (m)')
plt.xlim([-0.1,10])
plt.ylim([-2.5e12,2.5e12])
plt.title('Relationship between A and e for Constant Earth Lambda')
plt.show()
```

```python
t=numpy.linspace(0,4e8,1000)
omega1=1.98898e-7
r1=450000
r2=7.43e8
omega2=1.67699e-8
x1=r1*numpy.cos(omega1*t)
y1=r1*numpy.sin(omega1*t)
x2=r2*numpy.cos(omega2*t)
y2=r2*numpy.sin(omega2*t)
x3=x1+x2
y3=y1+y2
plt.plot(x1,y1,color='xkcd:blue',label='Earth Alone')
plt.plot(x2,y2,color='xkcd:red orange',label='Jupiter Alone')
plt.plot(x3,y3,color='xkcd:green',label='Both')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Comparison of Solar Motion for Systems with Earth and Jupiter')
plt.legend()
plt.show()
```
