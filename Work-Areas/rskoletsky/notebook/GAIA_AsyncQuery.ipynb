```python


from astroquery.gaia import Gaia
tables = Gaia.load_tables(only_names=True)
for table in (tables):
    print (table.get_qualified_name())

```

```python
job1 = Gaia.launch_job_async("SELECT * \
FROM gaiadr1.gaia_source AS g, gaiadr1.tmass_best_neighbour AS tbest, gaiadr1.tmass_original_valid AS tmass \
WHERE g.source_id = tbest.source_id AND tbest.tmass_oid = tmass.tmass_oid \
AND CONTAINS(POINT('ICRS',g.ra,g.dec),BOX('ICRS',68.0,25.0,.1,.11))=1 ;", dump_to_file=False)
```

```python
p = job1.get_results() 
print (p['phot_g_mean_mag', 'j_m', 'h_m', 'ks_m']) 
```

```python
plt.plot(p['ra'],p['dec'],'k.',markersize=10)
plt.show()
```

```python
job2 = Gaia.launch_job_async("SELECT * \
FROM gaiadr1.gaia_source AS g, gaiadr1.sdss_dr9_best_neighbour AS sdssbest, gaiadr1.sdssdr9_original_valid AS sdss \
WHERE g.source_id = sdssbest.source_id AND sdssbest.sdssdr9_oid = sdss.sdssdr9_oid \
AND CONTAINS(POINT('ICRS',g.ra,g.dec),BOX('ICRS',68.0,25.0,.1,.11))=1 ;", dump_to_file=False)
```
