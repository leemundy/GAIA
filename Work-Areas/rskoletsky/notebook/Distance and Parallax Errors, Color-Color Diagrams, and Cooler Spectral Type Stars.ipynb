```python
from astropy.table import Table
import matplotlib.pyplot as plt
import math
import numpy
import statistics
#Imports all relevant software for Python
Cat2=Table.read('TGAS_d125_Wright03_2Mass.vot')
#Defines Catalog 2 as this table, the last cross-match
print("The columns in Cat2 are: ")
#Gives text introduction to column names
print(Cat2.colnames)
#Shows the Catalog 2 columns for reference
parallaxerror=Cat2["e_Plx"]
#Defines the error in the parallax angle
parallax=Cat2["Plx"]
#Defines the parallax angle
plt.xlabel('Parallax (mas)')
plt.xlim(8,175)
plt.ylabel('Error in Parallax (mas)')
plt.ylim(0.2,1.02)
plt.title('Parallax Error versus Parallax')
plt.plot(parallax,parallaxerror,'ko',ms=1)
plt.show()
#Creates scatter plot of the error in parallax angle versus the parallax angle
parallax_arcseconds=(1/1000)*parallax
#Unit conversion
distance=1/parallax_arcseconds
#Defines distance in parsecs based on the parallax angle in arcseconds
distanceerror=numpy.divide(parallaxerror,(numpy.square(parallax)))
#Defines the error in distance in parsecs
plt.xlabel('Distance (pc)')
plt.xlim(0,125)
plt.ylabel('Error in Distance (pc)')
plt.ylim(-0.0001,0.0155)
plt.title('Distance Error versus Distance')
plt.plot(distance,distanceerror,'ko',ms=1)
plt.show()
#Creates scatter plot of the error in distance versus the distance
bins=30
#Makes 15 bins for histograms
plt.xlabel('Parallax Error (mas)')
plt.xlim(0.2,1)
plt.ylabel('Frequency')
plt.title('Histogram of Parallax Angle Errors')
plt.hist(parallaxerror,bins)
plt.show()
#Makes a histogram of the parallax angle errors with 15 bins
plt.xlabel('Distance Error (pc)')
plt.xlim(0,0.0155)
plt.ylabel('Frequency')
plt.title('Histogram of Distance Errors')
plt.hist(distanceerror,bins)
plt.show()
#Makes a histogram of the distance errors with 15 bins
Gapp=Cat2["__Gmag_"]
#Defines G apparent magnitude
Japp=Cat2["Jmag"]
#Defines J apparent magnitude
Happ=Cat2["Hmag"]
#Defines H apparent magnitude
Kapp=Cat2["Kmag"]
#Defines K apparent magnitude
J_Happ=Japp-Happ
#Defines apparent magnitude J-H
G_Japp=Gapp-Japp
#Defines apparent magnitude G-J
plt.xlabel('J-H')
plt.ylabel('G-J')
plt.title('G-J versus J-H Apparent Magnitude Color-Color Diagram')
plt.plot(J_Happ,G_Japp,'ko',ms=1)
plt.show()
#Makes an apparent magnitude color-color diagram with G-J versus J-H
Teff=Cat2["Teff"]
#Defines the effective temperature
SpType=Cat2["SpType"]
#Defines the spectral type
plt.xlabel('Effective Temperature (K)')
plt.ylabel('Frequency')
plt.title('Histogram of Effective Temperatures')
plt.hist(Teff,bins)
plt.show()
#Creates a histogram of effective temperatures in Kelvin
print(SpType)
bins=350
plt.xlabel('Spectral Type')
plt.ylabel('Frequency')
plt.title('Histogram of Spectral Types')
plt.hist(SpType,bins)
plt.show()
#Creates a histogram of stars' spectral types
```
