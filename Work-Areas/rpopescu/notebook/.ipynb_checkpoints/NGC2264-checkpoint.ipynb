```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
from astropy.table import Table
t = Table.read('Data/NGC2264_2.vot', format='votable')

print("Finished reading Table. Read ",len(t['__Gmag_'])," sources")
```

```python
print("The astropy reader loads the data into the following variable names:")
print(t.colnames)
```

```python
RA = t['RA_ICRS']
DE = t['DE_ICRS']
num = len(RA)
print(num)
```

```python
plt.scatter(RA,DE, s = 1)
plt.show()
```

```python
X1 = np.zeros(num)
Y1 = np.zeros(num)
for i in range(0,num):
    X1[i] = RA[i]
    Y1[i] = DE[i]
```

```python
from sklearn.neighbors import NearestNeighbors
X = np.column_stack((X1,Y1))
nbrs = NearestNeighbors(n_neighbors=5).fit(X)
a,b = nbrs.kneighbors(X)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
```
