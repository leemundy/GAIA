A notebook to test python functions before using those functions with real data in the other notebooks

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
import scipy.optimize
```

```python
#Testing to see how many points needed for covariance matrix

#number of points
num = 4

#generate sample data
x = np.linspace(5,10,num)
y = np.random.uniform(low=-1.6, high=-1.4, size=num)*x+20
y_err = np.ones(num)*3

#plot data and determine fit
plt.errorbar(x,y,yerr=y_err,marker = 'o',markersize=5,linestyle='--')
[m1,b1],V = np.polyfit(x,y,deg=1,w=1/y_err,cov=True)
alpha = -m1
alpha_err = V[0,0]
B_lin = m1*x+b1
line_label = r'$\alpha$ = '+ str(round(alpha,2)) + r' $\pm$ ' + str(round(alpha_err,2))
plt.plot(x,B_lin,label=line_label)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sample Linear Data and Fit')
plt.legend()
plt.show()
```

```python
#Test 100 datasets with varying error values to see if error in data affects error in coefficient

#number of data points
num = 5

for i in range(100):
    
    #generate sample data 
    x = np.linspace(5,10,num)
    y = 1.5*x+20
    y_err = np.random.uniform(low=0.1, high=1, size=num)

    #plot data and determine fit
    plt.errorbar(x,y,yerr=y_err,marker = 'o',markersize=5,linestyle='--')
    [m1,b1],V = np.polyfit(x,y,deg=1,w=1/y_err,cov=True)
    alpha = -m1
    alpha_err = V[0,0]
    B_lin = m1*x+b1
    line_label = r'$\alpha$ = '+ str(round(alpha,2)) + r' $\pm$ ' + str(round(alpha_err,2))
    plt.plot(x,B_lin,label=line_label)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Sample Linear Data and Fit')
    plt.legend()
    plt.show()
```

```python
#Test SCIPY OPTIMIZE CURVEFIT
#Test 100 datasets with varying error values to see if error in data affects error in coefficient

#number of data points
num = 5

#define function to be fitted
def f(x,m,b):
    
    y = m*x +b
    
    return y

for i in range(100):
    #generate data
    x = np.linspace(5,10,num)
    y = 1.5*x+20
    y_err = np.random.uniform(low=0.1, high=10, size=num)

    #plot data and determine fit
    plt.errorbar(x,y,yerr=y_err,marker = 'o',markersize=5,linestyle='--')
    [popt,pcov] = scipy.optimize.curve_fit(f,x,y,sigma=y_err,absolute_sigma=True)
    print(popt)
    print(pcov)
    alpha = -popt[0]
    alpha_err = pcov[0,0]
    B_lin = popt[0]*x+popt[1]
    line_label = r'$\alpha$ = '+ str(round(alpha,2)) + r' $\pm$ ' + str(round(alpha_err,2))
    plt.plot(x,B_lin,label=line_label)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Sample Linear Data and Fit')
    plt.legend()
    plt.show()
```
